```python
def rps_game ():
    # 'Play Rock Paper Scissors!'
    print('1-Rock, 2-Paper, 3-Scissors')

    playing = True
    while playing:
        # Ask for player plays using numbers
        p1 = input("\nPlayer 1, enter (1/2/3): ")
        p2 = input("Player 2, enter (1/2/3): ")

        # Compare the choices
        if p1 == p2:
            print("It's a tie!")

        # Player 1 wins
        elif (p1 == "1" and p2 == "3") or \
             (p1 == "3" and p2 == "2") or \
             (p1 == "2" and p2 == "1"):
            print("Congratulations Player 1! You won this round.")

        # Player 2 wins
        elif (p2 == "1" and p1 == "3") or \
             (p2 == "3" and p1 == "2") or \
             (p2 == "2" and p1 == "1"):
            print("Congratulations Player 2! You won this round.")

        else:
            print("Invalid input! Please use numbers 1, 2, or 3.")
            continue

        # Ask to start a new game
        again = input("Do you want to start a new game? (y/n): ")
        if again != "y":
            print("Thanks for playing!")
            playing = False

rps_game ()
```
